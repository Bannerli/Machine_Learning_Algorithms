```python
import autograd.numpy as np
from autograd import grad
import matplotlib.pyplot as plt

def gradient_descent(g, alpha, max_its, w):
    #compute the gradient of our input function - note this is a function tool
    gradient = grad(g)
    #define a list to record the history of w
    history = []
    #run the gradient descent loop
    best_w = w
    best_eval = g(w)
    for k in range(max_its):
        #evaluate the gradient
        grad_eval = gradient(w)
        #take gradient descent step
        w = w - alpha * grad_eval
        history.append(w)
        #return only the weighted providing the lowest evaluation
        test_eval = g(w)
        if test_eval < best_eval:
            best_eval = test_eval
            best_w = w
    return best_w, history

def gradient_descent_byhand(g, alpha, max_its, w):
    #define a list to record the history of w
    history = []
    #run the gradient descent loop
    best_w = w
    best_eval = g(w)
    for k in range(max_its):
        #evaluate the gradient
        grad_eval = g(w)
        #take gradient descent step
        w = w - alpha * grad_eval
        history.append(w)
        #return only the weighted providing the lowest evaluation
        test_eval = g(w)
        if test_eval < best_eval:
            best_eval = test_eval
            best_w = w
    return best_w, history

def costfunction_history(weight_history, g):
    #loop over weight history and compute associated cost function history
    cost_history = []
    for i in range(len(weight_history)):
        cost_history.append(g(weight_history[i]))
    fig, axes = plt.subplots(1, 1, figsize = (6, 3))
    axes.plot(range(len(weight_history)),cost_history)
    axes.set_xlabel("iterations")
    axes.set_ylabel("cost function")
    plt.show()
    
def costfunction_singelplot(w_history_1, w_history_2, w_history_3, g):
    #loop over weight history and compute associated cost function history
    cost_history_1 = []
    cost_history_2 = []
    cost_history_3 = []
    for i in range(len(w_history_1)):
        cost_history_1.append(g(w_history_1[i]))
        cost_history_2.append(g(w_history_2[i]))
        cost_history_3.append(g(w_history_3[i]))
    fig, axes = plt.subplots(1, 1, figsize = (6, 3))
    axes.plot(range(len(w_history_1)),cost_history_1, label = "alhpa = 1")
    axes.plot(range(len(w_history_2)),cost_history_2, label = "alpha = 0.1")
    axes.plot(range(len(w_history_3)),cost_history_3, label = "alpha = 0.01")
    axes.legend()
    axes.set_xlabel("iterations")
    axes.set_ylabel("cost function")
    plt.show()
    
# create the input function
g = lambda w: 1/float(50)*(w**4 + w**2 + 10*w)   
# try other functions too!  Like g = lambda w: np.cos(2*w) , g = lambda w: np.sin(5*w) + 0.1*w**2, g = lambda w: np.cos(5*w)*np.sin(w)
# run gradient descent
```

```python
# compute the derivatives by hand
def derivative(w):
    return 1/float(50)*(4*w**3 + 2*w + 10)
```

```python
# when alpha equal 1
weigh_best_1, weight_history_1 = gradient_descent_byhand(g = derivative,alpha = 1,max_its = 1000,w = 2.0)
# use cost_history plotter
costfunction_history(weight_history_1,g)
```

```python
# when alpha equal 0.1
weigh_best_2, weight_history_2 = gradient_descent_byhand(g = derivative,alpha = 10**-1,max_its = 1000,w = 2.0)
# use cost_history plotter
costfunction_history(weight_history_2,g)
```

```python
# when alpha equal 0.01
weigh_best_3, weight_history_3 = gradient_descent_byhand(g = derivative,alpha = 10**-2,max_its = 1000,w = 2.0)
# use cost_history plotter
costfunction_history(weight_history_3, g)
```

```python
# plot all curves in one plot
costfunction_singelplot(weight_history_1,  weight_history_2, weight_history_3, g)
```

When the steplength equals 0.1(alpha = 0.1), the curve shows the best property, 
which works best for the particular function and initial point.
